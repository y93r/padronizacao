# Padronização

- Criar funções para padronização dos campos de cidade e UF.
    - Cidade: remover espaços excedentes, remover acentos e aplicar formato capitalização de título.
      Exemplo: "são paulo" → "Sao Paulo".
    - UF: remover espaços excedentes, remover acentos e converter para letras maiúsculas.
      Exemplo: "sp" → "SP".

- Salvar o tratamento em um arquivo.

```python
import pandas as pd
import re
from unidecode import unidecode
```

```python
#Função para padronizar nomes de cidades
def padronizar_cidade(texto):
    if not texto:
        return None
    #Remove espaços do início e fim
    texto = texto.strip()
    #Remove acentos
    texto = unidecode(texto)
    #Remove espaços duplicados no meio
    texto = re.sub(r'\s+', ' ', texto)
    #Title Case: "Sao Paulo"
    texto = texto.title()

    return texto
```

```python
#Função para padronizar uf para maiúscula
def padronizar_uf(texto):
    if not texto:
        return None
    #Remove espaços do início e fim
    texto = texto.strip()
    #Remove acentos
    texto = unidecode(texto)
    #Todas maiúsculas
    texto = texto.upper()
    
    return texto
```

```python
#Ler o arquivo
df = pd.read_excel("cidades.xlsx")
display(df)
```

```python
#Aplicar as funções de padronização
df['cidade_destino'] = df['cidade_destino'].apply(padronizar_cidade)
df['uf_destino'] = df['uf_destino'].apply(padronizar_uf)
```

```python
display(df)
```

```python
#Salvar arquivo padronizado
df.to_excel('cidades_padronizadas.xlsx', index=False)
```
